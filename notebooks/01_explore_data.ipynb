```python
# Cell 1: Load libraries
import pandas as pd
import matplotlib.pyplot as plt

print("Libraries loaded!")
```

```python
# Cell 2: Load the training data
df_train = pd.read_csv(r'D:\GitHub_Works\NLP_Phonetic-Semantic-Mapping\data\processed\train.csv')
df_val = pd.read_csv(r'D:\GitHub_Works\NLP_Phonetic-Semantic-Mapping\data\processed\val.csv')
df_test = pd.read_csv(r'D:\GitHub_Works\NLP_Phonetic-Semantic-Mapping\data\processed\test.csv')

print("Data loaded successfully!")
print(f"Training samples: {len(df_train)}")
print(f"Validation samples: {len(df_val)}")
print(f"Test samples: {len(df_test)}")
```

```python
# Cell 3: See label distribution
print("Label Distribution in Training Data:")
print("-" * 30)
print(df_train['label'].value_counts())
```

```python
# Cell 4: View sample examples
print("Sample Examples from Each Category:")
print("=" * 50)

for label in ['positive', 'negative', 'neutral']:
    print(f"\n{label.upper()}:")
    print("-" * 30)
    samples = df_train[df_train['label'] == label]['text'].head(3)
    for i, text in enumerate(samples, 1):
        print(f"  {i}. {text}")
```

```python
# Cell: Create results folder
import os
os.makedirs('results', exist_ok=True)
print("Results folder ready!")
```

```python
# Cell 5: Create bar chart of label distribution
label_counts = df_train['label'].value_counts()

plt.figure(figsize=(8, 5))
plt.bar(label_counts.index, label_counts.values, color=['green', 'red', 'gray'])
plt.title('Sentiment Distribution in Training Data')
plt.xlabel('Sentiment Label')
plt.ylabel('Number of Samples')
plt.tight_layout()
plt.savefig('results/label_distribution.png')
plt.show()

print("Chart saved to: results/label_distribution.png")
```

```python
# Cell 6: Analyze text lengths
df_train['word_count'] = df_train['text'].apply(lambda x: len(x.split()))

print("Text Length Analysis:")
print("-" * 30)
print(f"Shortest text: {df_train['word_count'].min()} words")
print(f"Longest text: {df_train['word_count'].max()} words")
print(f"Average length: {df_train['word_count'].mean():.1f} words")
```

```python
# Cell 7: Word count by sentiment
print("Average Word Count by Sentiment:")
print("-" * 30)
print(df_train.groupby('label')['word_count'].mean().round(1))
```

```python
# Cell 8: Summary of Phase 1
print("=" * 50)
print("PHASE 1 COMPLETE - SUMMARY")
print("=" * 50)

print("\nDataset Overview:")
print(f"  - Training samples: {len(df_train)}")
print(f"  - Validation samples: {len(df_val)}")
print(f"  - Test samples: {len(df_test)}")
print(f"  - Total samples: {len(df_train) + len(df_val) + len(df_test)}")

print("\nLabel Distribution (Training):")
for label in ['positive', 'negative', 'neutral']:
    count = len(df_train[df_train['label'] == label])
    print(f"  - {label}: {count} samples")

print("\nText Statistics:")
print(f"  - Average length: {df_train['word_count'].mean():.1f} words")

print("\nFiles Created:")
print("  - data/processed/train.csv")
print("  - data/processed/val.csv")
print("  - data/processed/test.csv")
print("  - results/label_distribution.png")

print("\n" + "=" * 50)
print("Next is Phase 2")
print("=" * 50)
```
